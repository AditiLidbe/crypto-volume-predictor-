**FEATURE ENGINEERING**

```python
filtered_df = filtered_df.sort_values(by=['symbol', 'date']).reset_index(drop=True)
```

```python
filtered_df['liquidity_ratio'] = filtered_df['24h_volume'] / filtered_df['price']
```

```python
filtered_df['price_ma3'] = (
    filtered_df.groupby('symbol')['price']
    .transform(lambda x: x.rolling(window=3, min_periods=1).mean())
)

filtered_df['volatility3'] = (
    filtered_df.groupby('symbol')['price']
    .transform(lambda x: x.rolling(window=3, min_periods=1).std())
)
```

```python
filtered_df['price_pct_change'] = (
    filtered_df.groupby('symbol')['price']
    .pct_change()
)
```

```python
filtered_df['vpt'] = filtered_df['price_pct_change'] * filtered_df['24h_volume']
```

```python
filtered_df = filtered_df.dropna().reset_index(drop=True)
```

```python
filtered_df.head()
```

**MODEL** **SELECTION AND TRAINING**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


df = filtered_df.rename(columns={'24h_volume': 'volume'})

# Feature matrix
X = df[['price', 'volume', 'price_ma3', 'volatility3', 'price_pct_change', 'vpt']]

# Target variable
y = df['liquidity_ratio'] 

# Train-test split (80-20), no shuffle to simulate temporal order
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("R2 Score:", r2_score(y_test, y_pred))
print("RMSE:", rmse)
```

```python
import matplotlib.pyplot as plt

importances = model.feature_importances_
feature_names = X.columns

plt.barh(feature_names, importances)
plt.xlabel("Feature Importance")
plt.title("Random Forest Feature Importances")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20]
}
grid = GridSearchCV(RandomForestRegressor(), params, cv=3)
grid.fit(X_train, y_train)
print(grid.best_params_)
```
